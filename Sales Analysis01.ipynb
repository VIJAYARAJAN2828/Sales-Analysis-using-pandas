### Import Libraries

```python
import pandas as pd
import os
```

### Merging 12 Months of sales Data into a single file

```python
files=[file for file in os.listdir("Sales Data")]
all_months_data=pd.DataFrame()

for file in files:
    df=pd.read_csv('Sales Data/'+file)
    all_months_data=pd.concat([all_months_data,df])
    
all_months_data.to_csv('all_data.csv',index=False)
```

### Updated Dataframe

```python
all_data=pd.read_csv('all_data.csv')
all_data
```

### Clean the data

```python
all_data.isnull().sum()
```

```python
nan_data=all_data[all_data.isna().any(axis=1)]
nan_data.head()
```

```python
all_data=all_data.dropna(how='all')
all_data
```

#### Find 'or' and delete it 

```python
all_data=all_data[all_data['Order Date'].str[0]!='O']
all_data
```

### Add a month column which will be easy to read

```python
all_data.loc[:, 'Month'] = all_data['Month'].astype(int)
```

```python
all_data.dtypes
```

```python
all_data.loc[:, 'Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
all_data.loc[:, 'Price Each'] = all_data['Price Each'].astype(float)
```

### Add a sales column

```python
all_data.loc[:, 'Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data
```

### Adding State column

```python
all_data.loc[:, 'State'] = all_data['Purchase Address'].str.split(',').str[2]
```

#### 1. What was the  best month sales and How much was earned that month

```python
best_month_sales=all_data.groupby(['Month'])['Sales'].sum().to_frame()
best_month_sales.sort_values(['Sales'],ascending=False)
```

#### 2. Which city has the highest number of sales

```python
all_data.loc[:, 'City'] = all_data['Purchase Address'].str.split(',').str[1]
```

```python
city_sales=all_data.groupby(['City'])['Sales'].sum().to_frame()
city_sales.sort_values(['Sales'],ascending=False)
```

#### 3. What time should we dissplay ads to maximize likelihood of customers buying product

```python
all_data.loc[:, 'Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.loc[:, 'Hours']=all_data['Order Date'].dt.hour
```

```python
all_data.loc[:, 'Min']=all_data['Order Date'].dt.minute
```

```python
best_time=all_data.groupby(['Hours'])['Hour'].count().to_frame()
```

```python
best_time.sort_values(['Hour'],ascending=False)
```
